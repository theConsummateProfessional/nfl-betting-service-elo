```python
import pandas as pd
import numpy as np 
from sklearn import linear_model
import sklearn.metrics as sm
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
X = df[['away_team_first_downs',
       'home_team_first_downs', 'home_team_rush_attempts',
       'away_team_rush_attempts', 'home_team_rush_yards',
       'away_team_rush_yards', 'home_team_rush_touchdowns',
       'away_team_rush_touchdowns', 'home_team_completions',
       'away_team_completions', 'home_team_pass_attempts',
       'away_team_pass_attempts', 'home_team_pass_yards',
       'away_team_pass_yards', 'home_team_pass_touchdowns',
       'away_team_pass_touchdowns', 'home_team_sacks', 'away_team_sacks',
       'home_team_yards', 'away_team_yards', 'home_team_turnovers',
       'away_team_turnovers', 'home_team_penalties', 'away_team_penalties',
       'home_team_penalty_yards', 'away_team_penalty_yards',
       'home_team_third_down_percentage', 'away_team_third_down_percentage',
       'home_team_time_of_possession', 'away_team_time_of_possession', 'home_team_elo', 'away_team_elo']].to_numpy()
y = df[['home_team_score', 'away_team_score']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=.2, shuffle=False)
regressor = linear_model.LinearRegression()
regressor.fit(x_train, y_train)
y_test_pred = regressor.predict(x_test)
```

```python
print("Linear regressor performance:")
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, y_test_pred), 2))
print("Mean squared error =", round(sm.mean_squared_error(y_test, y_test_pred), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(y_test, y_test_pred), 2))
print("R2 score =", round(sm.r2_score(y_test, y_test_pred), 2))
```
